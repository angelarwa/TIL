```python
import urllib.request 
# URL과 저장 경로 지정하기
url = "http://uta.pw/shodou/img/28/214.png"
savename = "test.png"
# 다운로드 
mem = urllib.request.urlopen(url).read()
# 파일로 저장하기 
with open(savename, mode="wb") as f:
    f.write(mem)
    print("저장되었습니다...!")
```

```python
#find() : 지정된 태그 한 개 추출

from bs4 import BeautifulSoup

html = """
<html><body>
  <h1 id="title">스크레이핑이란?</h1>
  <p id="body">웹 페이지를 분석하는 것</p>
  <p>원하는 부분을 추출하는 것</p>
</body></html>
"""

# HTML 분석하기 
soup = BeautifulSoup(html, 'html.parser')

# find() 메서드로 원하는 부분 추출하기 
title = soup.find(id="title")
body = soup.find(id="body")

# 텍스트 부분 출력하기
print("#title=" + title.string)
print("#body=" + body.string)
```

```python
#네이버 영화
from bs4 import BeautifulSoup
import re
req = requests.get('https://movie.naver.com/movie/point/af/list.nhn?page=1')
html =req.text
soup = BeautifulSoup(html, 'html.parser')
titles = soup.select('.movie')
points = soup.select('td.title > div > em')
reviews = soup.select('td.title')
#print(reviews[0])

movie_title=[]
movie_point=[]
movie_review=[]
for dom in titles :
     movie_title.append(dom.text)

for dom in points :
     movie_point.append(dom.text)

for dom in reviews :
    content=dom.contents[6]
    content=re.sub("[\n\t]","",content) #공백제거
    content=re.sub("신고","",content) 
    movie_review.append(content)
    
for i in range(len(movie_title)):
    print('영화제목:', movie_title[i])
    print('평점:', movie_point[i])
    print('리뷰:', movie_review[i])
    print('------------------------------------')
```

## 정적 웹 페이지 크롤링과 스크래핑 예제

```python
#https://movie.daum.net/moviedb/grade?movieId=3126&type=netizen&page=1
#5개 페이지에서 각 댓글의 평점과 댓글 내용을 추출하여 출력하는 웹 크롤링 소스를 구현
#단, 평점이 있더라도 댓글이 비어있는 것은 제외합니다.

from bs4 import BeautifulSoup
import re
req = requests.get('https://movie.daum.net/moviedb/grade?movieId=3126&type=netizen&page=1')
html =req.text
soup = BeautifulSoup(html, 'html.parser')
titles = soup.select('em.link_profile')
points = soup.select('em.emph_grade')
reviews = soup.select('p.desc_review')
print(reviews)

movie_title=[]
movie_point=[]
movie_review=[]
for dom in titles :
     movie_title.append(dom.text)

for dom in points :
     movie_point.append(dom.text)

for dom in reviews :
    content=dom.contents[6]
    content=re.sub("[\n\t]","",content) #공백제거
    content=re.sub("신고","",content) 
    movie_review.append(content)
    
for i in range(len(movie_title)):
    print('영화제목:', movie_title[i])
    print('평점:', movie_point[i])
    print('리뷰:', movie_review[i])
    print('------------------------------------')
```
